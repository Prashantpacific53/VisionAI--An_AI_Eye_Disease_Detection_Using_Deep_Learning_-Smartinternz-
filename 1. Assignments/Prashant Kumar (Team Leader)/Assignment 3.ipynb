```python
# âœ… Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Enable inline plotting for Jupyter Notebook
%matplotlib inline

# --- 1ï¸âƒ£ NumPy: Solving Exercises ---
print("ðŸ”¹ NumPy Exercises\n")

# 1ï¸âƒ£ Create an array of zeros with 5 elements
array1 = np.zeros(5)
print("Array of Zeros:", array1)

# 2ï¸âƒ£ Create an array of ones with shape (3,3)
array2 = np.ones((3,3))
print("\nArray of Ones:\n", array2)

# 3ï¸âƒ£ Generate a random array of shape (4,4)
array3 = np.random.rand(4,4)
print("\nRandom 4x4 Array:\n", array3)

# 4ï¸âƒ£ Get the shape and size of an array
print("\nShape of array3:", array3.shape)
print("Size of array3:", array3.size)

# 5ï¸âƒ£ Reshape a 1D array to 2D
array4 = np.arange(12).reshape(3,4)
print("\nReshaped 3x4 Array:\n", array4)


# --- 2ï¸âƒ£ Pandas: Solving Exercises ---
print("\nðŸ”¹ Pandas Exercises\n")

# 1ï¸âƒ£ Create a simple DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35], 'Salary': [50000, 60000, 70000]}
df = pd.DataFrame(data)
print("Simple DataFrame:\n", df)

# 2ï¸âƒ£ Display basic information about the DataFrame
print("\nDataFrame Info:")
print(df.info())

# 3ï¸âƒ£ Display statistical summary
print("\nStatistical Summary:")
print(df.describe())

# 4ï¸âƒ£ Add a new column (Tax = 10% of Salary)
df['Tax'] = df['Salary'] * 0.1
print("\nDataFrame with Tax Column:\n", df)

# 5ï¸âƒ£ Filter data (People with Salary > 55,000)
filtered_df = df[df['Salary'] > 55000]
print("\nFiltered Data (Salary > 55,000):\n", filtered_df)


# --- 3ï¸âƒ£ Matplotlib: Drawing 10 Meaningful Graphs ---
print("\nðŸ”¹ Matplotlib Graphs\n")

# 1ï¸âƒ£ Line Graph - Salary vs Age
ages = [25, 30, 35, 40, 45]
salaries = [50000, 60000, 70000, 80000, 90000]

plt.figure(figsize=(6,4))
plt.plot(ages, salaries, marker='o', linestyle='-', color='b', label="Salary Growth")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Salary Growth with Age")
plt.legend()
plt.show()

# 2ï¸âƒ£ Bar Chart - Employee Salary Comparison
names = ["Alice", "Bob", "Charlie"]
salaries = [50000, 60000, 70000]

plt.figure(figsize=(6,4))
plt.bar(names, salaries, color=['red', 'green', 'blue'])
plt.xlabel("Employees")
plt.ylabel("Salary")
plt.title("Salary Comparison")
plt.show()

# 3ï¸âƒ£ Pie Chart - Market Share
companies = ["Google", "Microsoft", "Amazon", "Apple"]
market_share = [30, 25, 20, 25]

plt.figure(figsize=(6,6))
plt.pie(market_share, labels=companies, autopct='%1.1f%%', colors=['gold', 'blue', 'green', 'red'])
plt.title("Market Share Distribution")
plt.show()


# --- 4ï¸âƒ£ Seaborn: Drawing 10 Meaningful Graphs ---
print("\nðŸ”¹ Seaborn Graphs\n")

# Load Sample Dataset
df = sns.load_dataset("tips")  # Built-in dataset about restaurant bills

# 1ï¸âƒ£ Scatter Plot - Total Bill vs Tip
plt.figure(figsize=(6,4))
sns.scatterplot(x="total_bill", y="tip", data=df, hue="sex")
plt.title("Total Bill vs Tip Amount")
plt.show()

# 2ï¸âƒ£ Box Plot - Distribution of Bill Amount
plt.figure(figsize=(6,4))
sns.boxplot(x="day", y="total_bill", data=df, palette="coolwarm")
plt.title("Bill Amount Distribution by Day")
plt.show()

# 3ï¸âƒ£ Heatmap - Correlation Matrix (FIXED)
plt.figure(figsize=(6,4))

# Select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=['number'])

sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

```
